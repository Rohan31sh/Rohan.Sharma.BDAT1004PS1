```python
// question 7

def bbb(filename):
    a = {}
    lines = []
    with open(filename, 'r') as f:
        lines = f.readlines()
        f.close()
    for line in lines:
        bg = line.split('=')[0]
        patient = int(line.split('=')[1])
        if patient == 0:
            print(f"There are no patients of blood type {bg}")
        elif patient == 1:
            print(f"There is one patient of blood type {bg}")
        else:
            print(f"There are {patient} patients of blood type {bg}")

    
a = bbb('bloodtype.txt')
print(a)
```

```python
// question 8

def curconv(currency,amount):
    with open('currencies.txt','r') as file:
        for line in file:
            if currency in line:
                usDollerAmount = float(line.split()[1]) 
                return usDollerAmount * amount
            

if __name__ == '__main__':
    currency = input("Enter currency name:")
    amount = int(input("Enter amount:"))
    print(curconv(currency,amount))
```

```python
// question 10

def frequency(word):
    lis = [0] * 26
    for char in word:
        if char in 'abcdefghijklmnopqrstuvwxyz':
            lis[ord(char)-ord('a')] += 1

    return lis

if __name__ == "__main__":
    a = frequency('The quick red fox got bored and went home.')
    print(a)
```
